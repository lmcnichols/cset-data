```r
library(ggplot2)
options(repr.plot.height=5)
```

```r
data = read.csv("../data/H-1B/processed.csv")
data = data[order(data$Employer),]
data
```

# Big 5 Combined

```r
ggplot(aggregate(quant ~ date, data, sum), aes(date, quant)) + geom_line()
```

# Big 5 Separated

```r
ggplot(data, aes(date, quant)) + geom_line(aes(color=Employer), size=1)
```
